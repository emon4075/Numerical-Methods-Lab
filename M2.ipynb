# Algorithms For Data Streams (Online Methods)

## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import percentileofscore
```

## The Running Average (Online Mean)

```python
M_n = 0
i = 1
mean_array = []
number_array = []

while True:
    if i > 10:
        break

    number = int(input())
    number_array.append(number)

    M_n = M_n + ((number - M_n) / i)
    mean_array.append(M_n)
    
    i += 1

    print(f'Current Mean After {number}: {M_n}')
```

```python
plt.figure(figsize=(6,4))
plt.scatter(number_array, mean_array, marker='x', color='red')
plt.plot(number_array, mean_array, ls=':')
plt.xlabel('Number Over the Time')
plt.ylabel('Change of Mean')
plt.title('Dynamic/Online Mean')
plt.grid(alpha=0.4)
plt.show()
```

## Mean Absolute Deviation (Welford's Algorithm)

```python
M_n = 0
MAD_n = 0
i = 1
number_array = []

while True:
    if i > 10:
        break

    number = int(input())
    number_array.append(number)

    M_n = M_n + ((number - M_n) / i)

    current_deviation = abs(number - M_n)
    MAD_n = MAD_n + ((current_deviation - MAD_n) / i)

    i += 1
true_mean_absolute_deviation = np.mean(
    abs(np.mean(number_array) - np.array(number_array))
)

print(f"Mean Absolute Deviation (Calculated): {MAD_n}")
print(f"Mean Absolute Deviation (True): {true_mean_absolute_deviation}")
```

## Percentile Ranking (Exam)

A percentile rank indicates the percentage of scores in a comparison group that fall at or below a specific score, showing an individual's relative performance.

$$\text{Percentile Rank} = \left( \frac{\text{Count of numbers strictly less than } X}{\text{Total Count of numbers}} \right) \times 100$$

```python
# Using Library Function
numbers = np.random.uniform(1, 20, 1000)
user_input = float(input())
rank = percentileofscore(numbers, user_input, kind="strict")

print(f"The Percentile Rank of {user_input}: {rank:.2f}%")
```

```python
# Manual Calculation
less_than_user = np.sum(numbers < user_input)
rank = (less_than_user / 1000) * 100

print(f"The Percentile Rank of {user_input}: {rank:.2f}%")
```
