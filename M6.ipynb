# Series Approximation + Visualization and Interpolation

## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.interpolate import interp1d, CubicSpline
```

## Maclaurin Series

$cos(x)$ <br>
`A Maclaurin Series is a Taylor Series that is Expanded at a = 0`

```python
def maclaurin_cos(x, n_terms):
    result = 0
    for n in range(n_terms + 1):
        term = (((-1) ** n) * (x ** (2 * n))) / math.factorial(2 * n)
        result += term
    return result


x_values = np.linspace(-10, 10, 200)

true_cos_values = np.cos(x_values)

n_terms = 10
approx_values = []

for x in x_values:
    approx_values.append(maclaurin_cos(x, n_terms))

plt.figure(figsize=(6, 4))
plt.plot(x_values, true_cos_values, c="red", ls="--", label="True Cos")
plt.plot(x_values, approx_values, c="green", ls="-.", label="Approximate Cos")
plt.axhline(y=0, color="k", alpha=0.4)
plt.title(f"Maclaurin Series For Cos(x) of {n_terms} Terms")
plt.ylabel("f(x)")
plt.xlabel("x")
plt.grid()
plt.legend()
plt.show()
```

## Interpolation

### Linear Interpolation

```python
x_data = np.array([0, 1, 2, 3, 4, 5])
y_data = np.array([0, 0.8, 0.9, 0.1, -0.8, -1.0])

linear_interp = interp1d(x_data, y_data, kind="linear")

x = np.linspace(0, 5, 200)

plt.figure(figsize=(6, 4))
plt.scatter(x_data, y_data, color="k", label="Data Points")
plt.plot(x, linear_interp(x), color="r", label="Linear Interpolation")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Linear Interpolation")
plt.grid(alpha=0.4)
plt.legend()
plt.show()
```

### Cubic Spline Interpolation

```python
x_data = np.array([0, 1, 2, 3, 4, 5])
y_data = np.array([0, 0.8, 0.9, 0.1, -0.8, -1.0])

linear_interp = CubicSpline(
    x_data,
    y_data,
)

x = np.linspace(0, 5, 200)

plt.figure(figsize=(6, 4))
plt.scatter(x_data, y_data, color="k", label="Data Points")
plt.plot(x, linear_interp(x), color="r", label="Cubic Spline Interpolation")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Cubic Interpolation")
plt.grid(alpha=0.4)
plt.legend()
plt.show()
```
